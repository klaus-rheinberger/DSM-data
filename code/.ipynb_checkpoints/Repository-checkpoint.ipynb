## Photovoltaic (PV)

- source: [Photovoltaic Geographical Information System (PVGIS)](https://re.jrc.ec.europa.eu/pvg_tools/en/tools.html)
- time frame: up to and including 2016
- sampling interval: 1 hour
- location: as we like
- [legal notice](https://ec.europa.eu/info/legal-notice_en): "... reuse is allowed, provided appropriate credit is given and changes are indicated."

```python
# plot of a sample day:

df = pd.read_pickle('../data/pv_df.pickle')  # resampled to 15 minutes

df.loc['2016-07-01':'2016-07-31','P'].plot(marker='', figsize=(12,5))
plt.ylabel('kW')
plt.grid(True)
```

## Electric Vehicles (EV)

- source: [My Electric Avenue](http://myelectricavenue.info/) monitored the usage of
over 200 EVs, cf. ~/forschung/data/My_Electric_Avenue/Data/My Electric Avenue Published Technical Data Overview.pdf
- time frame: Jan 2014 - Jun 2015
- sampling interval: 1 second, event based
- location: UK
- legal notice: "This data was collected as part of a customer funded Low Carbon Networks Fund project. A condition of this funding is that collected data is released for further work by other parties." cf. My Electric Avenue Published Technical Data Overview.pdf

```python
# import pickle:
df = pd.read_pickle('../data/ev_df.pickle')

# add TripDuration (h):
df['TripDuration (h)'] = (df["TripStopDateTime"] - 
                          df["TripStartDateTime"])/pd.Timedelta('1 hour')

df.info()
```

```python
df.head(3)
```

### All Participants

```python
# some histograms:

my_col = 'TripDuration (h)'
# my_col = 'TripDistance (km)'
# my_col = 'Power Consumption (kWh)'

df[my_col].hist(bins=100, legend=True, figsize=(7,5))
plt.xlabel(my_col);
```

```python
# group by participants:

my_df = df.groupby('ParticipantID').agg(**{
    'trips first'    :pd.NamedAgg(column='TripStartDateTime', aggfunc='min'),
    'trips last'     :pd.NamedAgg(column='TripStopDateTime' , aggfunc='max'),
    'trips dist (km)':pd.NamedAgg(column='TripDistance (km)', aggfunc='sum'),
    'trips dur (h)'  :pd.NamedAgg(column='TripDuration (h)' , aggfunc='sum'),
    'trips count'    :pd.NamedAgg(column='TripDuration (h)' , aggfunc='count') })

my_df['trips mean dist (km)'] = my_df['trips dist (km)']/my_df['trips count']
```

```python
qgrid.show_grid(my_df, precision=2)
```

### Selected Participant

```python
ind = df['ParticipantID'] == 'GC10'
my_df = df.loc[ind,:].sort_values(by="TripStartDateTime")
```

```python
my_df['TripDuration (h)'].hist(bins=100, legend=True, figsize=(7,5))
plt.xlabel('trip duration (h)');
```

```python
my_x = "TripDistance (km)"

my_y = "Power Consumption (kWh)"
# my_y = "TripDuration (h)"

plt.figure(figsize=(5,5))
plt.scatter(my_x, my_y, data=my_df)
plt.xlabel(my_x)
plt.ylabel(my_y)
plt.grid(True)
```

## Households (HH)

- source: [data.open-power-system-data.org](https://data.open-power-system-data.org/household_data/), 6 residential households, 3 industrial buildings, 2 public buildings
- time frame: Mar 2016 - Jan 2017
- sampling interval: 15 minutes
- location: southern Germany
- legal notice: Creative Commons Attribution-International, Attribution in Chicago author-date style should be given as follows: "Open Power System Data. 2020. Data Package Household Data. Version 2020-04-15. https://data.open-power-system-data.org/household_data/2020-04-15/. (Primary data from various sources, for a complete list see URL)." cf. ~/forschung/data/OPSD/Household_Data/README.md

```python
# residential buildings:

df = pd.read_pickle('../data/hh_df.pickle')
```

```python
my_start = '2016-03-01'
my_end   = '2017-02-01'

(df.loc[my_start:my_end,:].diff()*4).plot(
        grid=True, figsize=(12,12), subplots=True)
plt.ylabel('demand (kW)')
plt.title(my_col);
```

```python
my_col   = 'DE_KN_residential3'
my_start = '2016-03-01'
my_end   = '2016-03-07'

(df.loc[my_start:my_end,my_col].diff()*4).plot(
        grid=True, figsize=(12,5))
plt.ylabel('demand (kW)')
plt.title(my_col);
```
